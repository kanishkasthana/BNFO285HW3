<h3>Part A

Defining Function to get 1000 Samples of size N from the Geometric Distribution

```r
getSamples=function(n){
    #Getting 1000 Samples
    samples=sapply(rep(n,1000),rgeom,p=0.3)
    return(samples)
}
```

Maximum likelihood Estimator of P is N/(N+ Sum of all values in Trial).
Defining function to get that maximum likelihood estimate

```r
getEstimateofP=function(sample){
    pEstimate=length(sample)/(length(sample)+sum(sample))
}
```

Each Column is a Single trial such that there are 1000 Column in total for each of the following variables.
Each trial is of size 100,1000, or 10000

```r
n100=getSamples(100)
n1000=getSamples(1000)
n10000=getSamples(10000)
```

Getting distribution of estimates of p. 

```r
pEstimateForN100=apply(n100,2,getEstimateofP)
pEstimateForN1000=apply(n1000,2,getEstimateofP)
pEstimateForN10000=apply(n10000,2,getEstimateofP)
```

Plotting Histograms for the distribution of the Estimated values of P for each N

```r
hist(pEstimateForN100,50,xlab="Estimated Values of P", main="Distribution of Estimated Values of P for N=100")
hist(pEstimateForN1000,50,xlab="Estimated Values of P", main="Distribution of Estimated Values of P for N=1000")
hist(pEstimateForN10000,50,xlab="Estimated Values of P", main="Distribution of Estimated Values of P for N=10000")
```

```r
#Printing Standard Errors:
print(sd(pEstimateForN100))
print(sd(pEstimateForN1000))
print(sd(pEstimateForN10000))
```

Showing the Standard Error is proportional to 1/Sqrt(N) in each case. We can show this my multipling the Calculated Standard Errors by Sqrt(N). This should give us roughly the same answer for each Trial with different Ns.

```r
#All the following Values should be similar after multiplying by Squareroot of N
print(sd(pEstimateForN100)*sqrt(100))
print(sd(pEstimateForN1000)*sqrt(1000))
print(sd(pEstimateForN10000)*sqrt(10000))
```

```r
#Generating Data X from distribution:
X=rgeom(1000000,0.3)
```

<h3> Part C

Defining Function to resample 1000 samples of size N from the distribution X


```r
reSample=function(n){   
    #Getting 1000 resamples with replacement
    samples=sapply(rep(n,1000),function(n){
        s=sample(X,size = n, replace = TRUE)
        return(s)
    })
    
    return(samples)
}
```

Getting Resamples of Different N

```r
resampledN100=reSample(100)
resampledN1000=reSample(1000)
resampledN10000=reSample(10000)
```

Getting Distribution of Estimated Pstar of P

```r
pEstimateForReSampledN100=apply(resampledN100,2,getEstimateofP)
pEstimateForReSampledN1000=apply(resampledN1000,2,getEstimateofP)
pEstimateForReSampledN10000=apply(resampledN10000,2,getEstimateofP)
```

Plotting Histograms for the resampled distributions of the Estimated values of P for each N

```r
hist(pEstimateForReSampledN100,50,xlab="Estimated Values of P", main="Resampled Distribution of Estimated Values of P for N=100")
hist(pEstimateForReSampledN1000,50,xlab="Estimated Values of P", main="Resampled Distribution of Estimated Values of P for N=1000")
hist(pEstimateForReSampledN10000,50,,xlab="Estimated Values of P", main="Resampled Distribution of Estimated Values of P for N=10000")
```

```r
#Printing Standard Errors:
print(sd(pEstimateForReSampledN100))
print(sd(pEstimateForReSampledN1000))
print(sd(pEstimateForReSampledN10000))
```
