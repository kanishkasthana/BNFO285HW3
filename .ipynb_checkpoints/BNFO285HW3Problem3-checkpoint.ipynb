Reading in data for homework

```r
data=read.csv("BNFO285_HW3_P3.csv")
```

```r
X=data["x"]
X=as.numeric(unlist(X))
#Plotting to see how X looks like
hist(X,100)
```

To calculate AIC for each distribution the Likelihood of each Distribution for the MLE of the parameters needs to be found.

```r
#Loading Library MASS for parameter calculation
library("MASS")
```

```r
poissonEstimate=fitdistr(X,densfun = "Poisson")
print("Poisson Estimate:")
print(poissonEstimate$estimate)
print(poissonEstimate$loglik)
geometricEstimate=fitdistr(X,densfun= "geometric")
print("Geometric Estimate:")
print(geometricEstimate$estimate)
print(geometricEstimate$loglik)
negativeBinomialEstimate=fitdistr(X,densfun = "negative binomial")
print("Negative Binomial Estimate:")
print(negativeBinomialEstimate$estimate)
print(negativeBinomialEstimate$loglik)
#Using Normal Approximation for Binomial Distribution:
binomialEstimateNormalApproximation=fitdistr(X,densfun = "normal")
print("Using Normal Approximation for Binomial Distribution:")
print(binomialEstimateNormalApproximation$estimate)
print(binomialEstimateNormalApproximation$loglik)
```
